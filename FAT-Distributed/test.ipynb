```python
import torch
import torch.nn as nn
```

```python
model1 = nn.Linear(10, 5)
params1 = [param.data.flatten() for param in model1.parameters()]
flatten_params1 = torch.cat(params1)
flatten_params1
```

```python
model2 = nn.Linear(10, 5)
params2 = [param.data.flatten() for param in model2.parameters()]
flatten_params2 = torch.cat(params2)
flatten_params2
```

```python
gather = [flatten_params1,flatten_params2]
gather = torch.stack(gather)
mean_gathered_params = torch.mean(gather, dim=0)
gather
```
